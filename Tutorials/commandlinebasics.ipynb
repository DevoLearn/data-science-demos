1.pwd : It gives us the absolute path

````
!pwd
````

2. ls Use the "ls" command to know what files are in the directory you are in

````
!ls
````

3. cd  Use the "cd" command to go to a directory

````
!cd sample_data
````

 4.mkdir  is used to create a directory  

````
!mkdir devolearn
````

5.touch  The touch command is used to create a file

````
!touch hellofriend.txt >> "devolean welcomes you"
````

 6.rm - Use the rm command to delete files and directories

````
!rm filename 
````

7.locate  The locate command is used to locate a file in a Linux 
System(this command first has to be installed)


```

```
